```python
# Import the necessary library
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

FOLDER_PATH = '/Users/matteom/shared-folder/Tesi/'  # Replace with the actual folder path

# Load the first CSV table
file_path_1 = FOLDER_PATH+'Automatic-PlotOverLineData.csv'  # Replace with the actual file path or file name
table1 = pd.read_csv(file_path_1)

# Load the second CSV table
file_path_2 = FOLDER_PATH+'Manual-PlotOverLineData.csv'  # Replace with the actual file path or file name
table2 = pd.read_csv(file_path_2)

table1 = table1.dropna()
table2 = table2.dropna()

# Compute the 'r' column as r = sqrt(x^2 + y^2) for table1
table1['r'] = np.sqrt(table1['x']**2 + table1['y']**2)

# Compute the 'r' column as r = sqrt(x^2 + y^2) for table2
table2['r'] = np.sqrt(table2['x']**2 + table2['y']**2)

# Display the first few rows of each table to verify the data is loaded
print("Automatic BCs CSV Table:")
display(table1.head())  # Displays the first few rows of table 1

print("\n Manual Lifting CSV Table:")
display(table2.head())  # Displays the first few rows of table 2
```

```python
# Create a plot of E_r versus r for both tables
plt.figure(figsize=(10, 6))

# Plot for table1
plt.plot(table1['r'], table1['E_r'], label='AutomaticBC', color='blue', marker='', linestyle='-')

# Plot for table2
plt.plot(table2['r'], table2['E_r'], label='ManualLifting', color='red', marker='', linestyle='--')

# Add labels and title
plt.xlabel('r', fontsize=12)
plt.ylabel('E_r', fontsize=12)
plt.title('E_r for both methods', fontsize=14)

# Add legend
plt.legend()

# Display the plot
plt.grid(True)
plt.show()
```
