```python
# Load both CSV files
file_1 = "../results/GlobRefQ2errore_punto_L2_H1-convergence_data.csv"

data_1 = pd.read_csv(file_1)


cycles = len(data_1['num_cells'])
h = np.zeros(cycles)
for i in range(cycles):
  h[i]=0.0001**(i+1)

# First subplot
plt.subplot(1, 2, 1)
plt.loglog(h, data_1['L2_errors'], 'r-o', label='L2 norm')
plt.loglog(h, h** 2 , 'g:', label="1/N^2")
plt.loglog(h, h , 'k:', label="1/N")
plt.xlabel("Number of Cells")
plt.ylabel("Error")
plt.legend()
plt.grid(True)

# Second subplot
plt.subplot(1, 2, 2)
plt.loglog(h, data_1['H1_errors'], 'b-o', label='H1 norm')
plt.loglog(h, h , 'k:', label="1/N")
plt.xlabel("Number of Cells")
plt.ylabel("Error")
plt.legend()
plt.grid(True)

# Adjust layout and show plot
plt.tight_layout()
plt.show()

print(len(data_1['num_cells']))
```
